```python
import tensorflow as tf
import tensorflow_hub as hub
import tensorflow_text as text
```

```python
encoder_url = 'https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4'
preprocess_url = 'https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3'
```

```python
bert_preprocess_model = hub.KerasLayer(preprocess_url)
#Creating a function pointer
```

```python
text_test = ['nice movie indeed', 'I love python programming']
text_preprocessed = bert_preprocess_model(text_test)
text_preprocessed.keys()
```

```python
text_preprocessed['input_mask']


#Tokens: CLS  nice movie indeed SEP (spatial token Separator separates two sentences)
```

```python
text_preprocessed['input_word_ids']

#Tokens: CLS  nice movie indeed SEP
```

```python
bert_model = hub.KerasLayer(encoder_url)

bert_results = bert_model(text_preprocessed)
bert_results.keys()
```

```python
bert_results['pooled_output']
```

The 768 vectors accurately represents full sentence.

```python
bert_results['sequence_output']
```

2 sentences; For each sentence there will be some padding;

```python
len(bert_results['encoder_outputs'])
```

12 encoders BERT base, each word has 128 words including padding; 768 embedding vector for each word
